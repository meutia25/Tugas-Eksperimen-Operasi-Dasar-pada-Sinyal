## Bagian A: Operasi pada Sinyal 1D

### A.1 Membuat Sinyal Diskrit

Kita akan membuat dua sinyal diskrit:
1.  **Sinyal pertama:** `x1[n] = sin(0.5πn)`
2.  **Sinyal kedua:** `x2[n] = 0` untuk `n < 5`, dan `x2[n] = 1` untuk `n ≥ 5`

Gunakan rentang `n = 0, 1, 2, ..., 30`.

```python
import numpy as np
import matplotlib.pyplot as plt

# Menentukan rentang nilai n
n = np.arange(0, 31)  # n = 0, 1, 2, ..., 30

# --- Sinyal x1[n] = sin(0.5πn) ---
x1 = np.sin(0.5 * np.pi * n)

# --- Sinyal x2[n] (Mirip Fungsi Unit Step) ---
# Bernilai 0 untuk n < 5 dan bernilai 1 untuk n >= 5
x2 = np.where(n >= 5, 1, 0)

# Menampilkan grafik sinyal x1[n]
plt.figure(figsize=(12, 5))
plt.stem(n, x1)
plt.title('Sinyal Diskrit x1[n] = sin(0.5πn)')
plt.xlabel('n')
plt.ylabel('x1[n]')
plt.grid(True)

# Menampilkan label sumbu x setiap 5 satuan agar lebih mudah dibaca
plt.xticks(n[::5])

plt.tight_layout()
plt.show()

# Menampilkan grafik sinyal x2[n]
plt.figure(figsize=(12, 5))
plt.stem(n, x2)
plt.title('Sinyal Diskrit x2[n] (Mirip Fungsi Unit Step)')
plt.xlabel('n')
plt.ylabel('x2[n]')
plt.grid(True)

# Menampilkan label sumbu x setiap 5 satuan agar lebih mudah dibaca
plt.xticks(n[::5])

plt.tight_layout()
plt.show()
```

### Penjelasan Karakteristik Sinyal

#### Sinyal x1[n] = sin(0.5πn)
Sinyal ini adalah **sinyal sinusoidal diskrit**. Karakteristik utamanya adalah:
-   **Periodik:** Karena `sin(x)` adalah fungsi periodik, `x1[n]` juga periodik. Dengan frekuensi `0.5π`, periodenya adalah `2π / (0.5π) = 4` sampel. Ini berarti sinyal akan mengulang bentuknya setiap 4 nilai `n`.
-   **Amplitudo:** Amplitudo maksimumnya adalah 1 dan minimumnya adalah -1.
-   **Kontinyu vs Diskrit:** Meskipun `sin(x)` adalah fungsi kontinyu, karena kita mengambil sampel hanya pada nilai `n` integer, ini menjadi sinyal diskrit.

#### Sinyal x2[n] = {0, n < 5; 1, n ≥ 5}
Sinyal ini merupakan bentuk dari **sinyal *unit step* yang digeser (shifted unit step)**. Karakteristik utamanya adalah:
-   **Amplitudo Konstan:** Sinyal ini memiliki amplitudo 0 untuk nilai `n` kurang dari 5, dan kemudian melompat ke amplitudo konstan 1 untuk `n` sama dengan atau lebih besar dari 5.
-   **Non-Periodik:** Sinyal ini tidak berulang dan memiliki perubahan diskrit pada `n=5`.
-   **Representasi Digital:** Sinyal seperti ini sering digunakan dalam sistem digital untuk merepresentasikan awal atau akhir suatu peristiwa, atau sebagai input dasar.

### A.2 Operasi Penjumlahan Sinyal

Kita akan melakukan operasi penjumlahan sinyal: `y[n] = x1[n] + x2[n]`.

```python
import matplotlib.pyplot as plt
import numpy as np

# Diasumsikan n, x1, dan x2 sudah didefinisikan sebelumnya
# Jika program dijalankan secara terpisah, aktifkan kode berikut:
# n = np.arange(0, 31)
# x1 = np.sin(0.5 * np.pi * n)
# x2 = np.where(n >= 5, 1, 0)

# Melakukan operasi penjumlahan dua sinyal
y = x1 + x2

# Menampilkan grafik x1[n], x2[n], dan y[n]
plt.figure(figsize=(15, 10))

# Grafik sinyal x1[n]
plt.subplot(3, 1, 1)
plt.stem(n, x1)
plt.title('Sinyal x1[n] = sin(0.5πn)')
plt.xlabel('n')
plt.ylabel('x1[n]')
plt.grid(True)

# Menampilkan label sumbu x setiap 5 satuan
plt.xticks(n[::5])

# Grafik sinyal x2[n]
plt.subplot(3, 1, 2)
plt.stem(n, x2)
plt.title('Sinyal x2[n] (Mirip Fungsi Unit Step)')
plt.xlabel('n')
plt.ylabel('x2[n]')
plt.grid(True)

# Menampilkan label sumbu x setiap 5 satuan
plt.xticks(n[::5])

# Grafik hasil penjumlahan y[n]
plt.subplot(3, 1, 3)
plt.stem(n, y)
plt.title('Sinyal Hasil Penjumlahan y[n] = x1[n] + x2[n]')
plt.xlabel('n')
plt.ylabel('y[n]')
plt.grid(True)

# Menampilkan label sumbu x setiap 5 satuan
plt.xticks(n[::5])

# Mengatur tata letak agar grafik tidak saling bertumpuk
plt.tight_layout()

# Menampilkan seluruh grafik
plt.show()
```

### Analisis Perubahan Bentuk Sinyal Setelah Dijumlahkan

Setelah operasi penjumlahan `y[n] = x1[n] + x2[n]`, sinyal `y[n]` menunjukkan kombinasi karakteristik dari kedua sinyal asalnya. Untuk `n < 5`, `x2[n]` bernilai 0, sehingga `y[n]` identik dengan `x1[n]`. Untuk `n ≥ 5`, `x2[n]` bernilai 1, yang berarti `y[n]` menjadi `x1[n] + 1`. Ini menyebabkan sinyal `x1[n]` terangkat ke atas (tergeser secara vertikal) sebesar 1 unit pada rentang `n ≥ 5`, sementara bentuk sinusoidalnya tetap dipertahankan.

### Pertanyaan Analisis

1.  **Apa yang terjadi pada amplitudo sinyal setelah dilakukan penjumlahan?**
    Amplitudo sinyal hasil penjumlahan `y[n]` secara lokal berubah. Untuk `n < 5`, amplitudo `y[n]` sama dengan `x1[n]`. Untuk `n ≥ 5`, amplitudo `y[n]` meningkat sebesar 1 dibandingkan dengan `x1[n]` pada setiap titik `n`. Nilai puncak `y[n]` bisa mencapai `1 + 1 = 2` (saat `x1[n]=1` dan `x2[n]=1`) dan nilai minimum bisa `(-1) + 1 = 0` (saat `x1[n]=-1` dan `x2[n]=1`). Jadi, rentang amplitudo secara keseluruhan dapat bertambah.

2.  **Apakah bentuk sinyal hasil penjumlahan masih menyerupai salah satu sinyal asal?**
    Ya, sinyal hasil penjumlahan `y[n]` masih sangat menyerupai sinyal `x1[n]`, terutama pada bagian di mana `x2[n]` adalah konstan (baik 0 maupun 1). Perbedaannya adalah 'level DC' atau nilai rata-rata sinyal `x1[n]` terangkat sebesar 1 unit pada bagian `n ≥ 5` karena pengaruh `x2[n]`. Namun, karakteristik bentuk gelombang sinusoidal dari `x1[n]` tetap terlihat jelas.

3.  **Dalam kasus nyata, operasi penjumlahan sinyal dapat digunakan untuk apa?**
    Dalam kasus nyata, operasi penjumlahan sinyal memiliki banyak aplikasi, antara lain:
    *   **Audio Mixing:** Menggabungkan beberapa trek audio (misalnya vokal, instrumen) menjadi satu sinyal audio komposit.
    *   **Sensor Fusion:** Menggabungkan data dari beberapa sensor untuk mendapatkan informasi yang lebih akurat atau lengkap. Contohnya, menggabungkan sinyal dari sensor suhu dan kelembaban.
    *   **Noise Reduction:** Jika kita memiliki sinyal yang sama diulang beberapa kali dengan noise yang berbeda, menjumlahkan (atau merata-ratakan) sinyal-sinyal tersebut dapat membantu mengurangi noise acak dan menonjolkan sinyal asli.
    *   **Image Blending:** Menggabungkan dua gambar atau lebih (misalnya, gambar latar belakang dengan objek) dengan menambahkan nilai pikselnya, seringkali dengan bobot tertentu, untuk menciptakan gambar baru. Ini adalah dasar dari banyak efek visual.
    *   **Superposisi Sinyal Komunikasi:** Dalam sistem komunikasi, beberapa sinyal informasi dapat dijumlahkan (dimultiplex) untuk ditransmisikan melalui satu saluran.

### A.3 Operasi Penggeseran Sinyal

Kita akan melakukan operasi penggeseran sinyal: `y[n] = x[n − k]`.
Kita akan menggunakan sinyal `x1[n]` sebagai sinyal asli dan menggesernya dengan minimal tiga nilai `k` yang berbeda: `k = -3`, `k = 0`, dan `k = 4`.

```python
import matplotlib.pyplot as plt
import numpy as np

# Diasumsikan n dan x1 sudah didefinisikan sebelumnya
# Jika program dijalankan secara terpisah, aktifkan kode berikut:
# n = np.arange(0, 31)
# x1 = np.sin(0.5 * np.pi * n)

# Menentukan nilai pergeseran (k)
k_values = [-3, 0, 4]

# Menyimpan hasil sinyal yang telah digeser
y_shifted = {}

# Fungsi untuk melakukan pergeseran sinyal
def signal_shift(signal, k):
    shifted_signal = np.zeros_like(signal, dtype=float)

    if k >= 0:
        # Pergeseran ke kanan (delay)
        shifted_signal[k:] = signal[:-k] if k > 0 else signal
    else:
        # Pergeseran ke kiri (advance)
        shifted_signal[:k] = signal[-k:]

    return shifted_signal

# Membuat sinyal hasil pergeseran
for k in k_values:

    # Membuat array kosong dengan ukuran yang sama seperti x1
    y_n = np.zeros_like(x1, dtype=float)

    # Melakukan pergeseran berdasarkan persamaan y[n] = x[n-k]
    for i in range(len(n)):
        idx = i - k

        # Memastikan indeks masih berada dalam batas array
        if 0 <= idx < len(n):
            y_n[i] = x1[idx]

    # Menyimpan hasil pergeseran
    y_shifted[k] = y_n

# Menampilkan grafik terpisah

plt.figure(figsize=(15, 12))

# Grafik sinyal asli
plt.subplot(len(k_values) + 1, 1, 1)
plt.stem(n, x1)
plt.title('Sinyal Asli x1[n]')
plt.xlabel('n')
plt.ylabel('x1[n]')
plt.grid(True)
plt.xticks(n[::5])

# Grafik sinyal hasil pergeseran
for i, k in enumerate(k_values):
    plt.subplot(len(k_values) + 1, 1, i + 2)
    plt.stem(n, y_shifted[k])

    plt.title(f'Sinyal Hasil Penggeseran y[n] = x1[n - {k}] (k = {k})')
    plt.xlabel('n')
    plt.ylabel('y[n]')
    plt.grid(True)
    plt.xticks(n[::5])

# Mengatur tata letak grafik
plt.tight_layout()
plt.show()

# Menampilkan perbandingan dalam satu grafik

plt.figure(figsize=(15, 6))

# Grafik sinyal asli
plt.stem(
    n, x1,
    linefmt='b-',
    markerfmt='bo',
    label='Sinyal Asli x1[n]'
)

# Warna untuk masing-masing hasil pergeseran
colors = ['g', 'r', 'c', 'm', 'y', 'k']

# Grafik sinyal hasil pergeseran
for i, k in enumerate(k_values):
    plt.stem(
        n,
        y_shifted[k],
        linefmt=f'{colors[i]}--',
        markerfmt=f'{colors[i]}o',
        label=f'x1[n - {k}]'
    )

plt.title('Perbandingan Sinyal Asli dan Hasil Penggeseran')
plt.xlabel('n')
plt.ylabel('Amplitudo')
plt.grid(True)
plt.xticks(n[::5])
plt.legend()

# Mengatur tata letak grafik
plt.tight_layout()
plt.show()
```

### Pertanyaan Analisis

1.  **Apa perbedaan efek k positif dan k negatif?**
    *   **`k` positif (misalnya `k = 4`):** Menggeser sinyal ke kanan (delay). Ini berarti nilai sinyal yang tadinya muncul di `n` sekarang akan muncul di `n + k`. Sinyal terlihat 'terlambat' karena membutuhkan waktu lebih lama untuk mencapai posisi tertentu. Bagian awal sinyal (`n < k`) akan menjadi nol atau tidak terdefinisi (tergantung implementasi, dalam kasus ini diisi nol).
    *   **`k` negatif (misalnya `k = -3`):** Menggeser sinyal ke kiri (advance). Ini berarti nilai sinyal yang tadinya muncul di `n` sekarang akan muncul di `n - |k|`. Sinyal terlihat 'mendahului' karena muncul lebih awal. Bagian akhir sinyal (`n > n_max - |k|`) akan menjadi nol atau tidak terdefinisi.

2.  **Bagaimana penggeseran sinyal dapat digunakan untuk simulasi delay?**
    Penggeseran sinyal ke kanan (yaitu, `k` positif pada `x[n-k]`) secara langsung mensimulasikan *delay* atau penundaan. Jika `x[n]` adalah sinyal input, maka `y[n] = x[n-k]` adalah sinyal output yang sama dengan input, tetapi muncul `k` sampel lebih lambat. Ini adalah prinsip dasar dalam banyak aplikasi, seperti simulasi gema pada audio, penundaan dalam transmisi data, atau modeling waktu propagasi sinyal dalam sistem fisik.

3.  **Mengapa *time alignment* penting dalam pengolahan sinyal?**
    *Time alignment* (penyelarasan waktu) sangat penting dalam pengolahan sinyal karena beberapa alasan:
    *   **Koreksi Fase:** Dalam sistem di mana sinyal dari berbagai sumber harus diinterferensi secara konstruktif atau dianalisis hubungannya, penyelarasan waktu memastikan bahwa puncak dan lembah sinyal cocok. Ketidakselarasan dapat menyebabkan pembatalan sinyal atau kesalahan pengukuran.
    *   **Korelasi dan Deteksi:** Banyak algoritma deteksi dan korelasi sinyal mengandalkan perbandingan bentuk sinyal pada waktu yang tepat. Jika sinyal tidak selaras, korelasi dapat rendah dan deteksi menjadi tidak akurat.
    *   **Fusi Data:** Dalam sistem multisensor, data dari sensor yang berbeda mungkin tiba pada waktu yang sedikit berbeda. Penyelarasan waktu memastikan bahwa semua data yang digabungkan merujuk pada momen yang sama di dunia nyata, memungkinkan interpretasi yang akurat.
    *   **Kualitas Audio/Video:** Dalam produksi multimedia, penyelarasan waktu antara audio dan video sangat penting untuk pengalaman pengguna yang baik. Ketidakselarasan sekecil apa pun dapat mengganggu.
    *   **Filtering Adaptif:** Dalam filter adaptif, sinyal referensi harus selaras dengan sinyal yang ingin diproses untuk performa optimal.

### A.4 Operasi Amplifikasi Sinyal

Kita akan melakukan operasi amplifikasi sinyal: `y[n] = αx[n]`.
Kita akan menggunakan sinyal `x1[n]` sebagai sinyal asli dan mengamplifikasinya dengan minimal empat nilai `α`: `0.5`, `1`, `2`, dan `-1`.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Diasumsikan n dan x1 sudah didefinisikan sebelumnya
# Jika program dijalankan secara terpisah, aktifkan kode berikut:
# n = np.arange(0, 31)
# x1 = np.sin(0.5 * np.pi * n)

# Menentukan nilai faktor pengali (α)
alpha_values = [0.5, 1, 2, -1]

# Menyimpan sinyal hasil amplifikasi
y_amplified = {}

# Membuat sinyal hasil amplifikasi
for alpha in alpha_values:
    y_amplified[alpha] = alpha * x1

# Menampilkan grafik

plt.figure(figsize=(15, 15))

# Grafik sinyal asli
plt.subplot(len(alpha_values) + 1, 1, 1)
plt.stem(n, x1)
plt.title('Sinyal Asli x1[n]')
plt.xlabel('n')
plt.ylabel('x1[n]')
plt.grid(True)
plt.xticks(n[::5])

# Menetapkan batas sumbu y agar semua grafik mudah dibandingkan
plt.ylim(-2.5, 2.5)

# Grafik hasil amplifikasi
for i, alpha in enumerate(alpha_values):
    plt.subplot(len(alpha_values) + 1, 1, i + 2)
    plt.stem(n, y_amplified[alpha])

    plt.title(f'Sinyal Hasil Amplifikasi y[n] = {alpha} × x1[n] (α = {alpha})')
    plt.xlabel('n')
    plt.ylabel('y[n]')
    plt.grid(True)
    plt.xticks(n[::5])

    # Menetapkan batas sumbu y yang sama
    plt.ylim(-2.5, 2.5)

# Mengatur tata letak grafik
plt.tight_layout()
plt.show()

# Membuat tabel perbandingan efek α

comparison_data = []

for alpha in alpha_values:

    if alpha > 1:
        effect = 'Amplitudo sinyal membesar (diperkuat).'

    elif 0 < alpha < 1:
        effect = 'Amplitudo sinyal mengecil (dilemahkan).'

    elif alpha == 1:
        effect = 'Tidak ada perubahan pada amplitudo sinyal.'

    elif alpha == -1:
        effect = 'Amplitudo tetap, tetapi polaritas sinyal terbalik.'

    elif alpha < 0:
        effect = 'Amplitudo membesar dan polaritas sinyal terbalik.'

    elif alpha == 0:
        effect = 'Sinyal menjadi nol (hilang sepenuhnya).'

    comparison_data.append({
        'Nilai α': alpha,
        'Efek terhadap Sinyal': effect
    })

# Mengubah data menjadi tabel
comparison_df = pd.DataFrame(comparison_data)

# Menampilkan tabel
print("Tabel Perbandingan Efek Setiap Nilai α:")
print(comparison_df.to_markdown(index=False))
```

### Pertanyaan Analisis

1.  **Apa yang terjadi ketika α > 1?**
    Ketika `α > 1`, sinyal mengalami **pembesaran amplitudo (amplification)**. Setiap nilai sampel sinyal asli dikalikan dengan `α`, sehingga amplitudo sinyal hasil amplifikasi menjadi lebih besar dari sinyal asli. Sinyal menjadi lebih 'kuat' atau 'intens'.

2.  **Apa yang terjadi ketika 0 < α < 1?**
    Ketika `0 < α < 1`, sinyal mengalami **pengecilan amplitudo (attenuation)**. Setiap nilai sampel sinyal asli dikalikan dengan `α`, sehingga amplitudo sinyal hasil amplifikasi menjadi lebih kecil dari sinyal asli. Sinyal menjadi lebih 'lemah' atau 'kurang intens'.

3.  **Apa yang terjadi ketika α bernilai negatif?**
    Ketika `α` bernilai negatif, sinyal mengalami **pembalikan polaritas (phase inversion)**. Selain itu, jika `|α| > 1`, amplitudo sinyal membesar; jika `0 < |α| < 1`, amplitudo mengecil; dan jika `α = -1`, amplitudo tetap sama tetapi sinyal terbalik sepenuhnya (fase bergeser 180 derajat).

4.  **Bagaimana konsep amplifikasi ini berkaitan dengan *gain* pada sistem audio?**
    Konsep amplifikasi ini secara langsung berkaitan dengan *gain* pada sistem audio. *Gain* adalah faktor pengali yang diterapkan pada amplitudo sinyal audio. Pengaturan *gain* pada *mixer*, *amplifier*, atau perangkat audio lainnya bertujuan untuk:
    *   **Meningkatkan volume:** Jika sinyal input terlalu lemah (misalnya dari mikrofon), *gain* yang positif (`α > 1`) akan memperkuat sinyal sehingga dapat didengar dengan jelas.
    *   **Mengurangi volume:** Jika sinyal input terlalu kuat, *gain* yang `0 < α < 1` (sering disebut *pad* atau *trim* dalam audio) dapat digunakan untuk melemahkan sinyal agar tidak terjadi *clipping* atau distorsi.
    *   **Inversi fase:** Dalam beberapa kasus, *gain* negatif (`α = -1`) dapat digunakan untuk membalik polaritas sinyal, yang berguna untuk mengatasi masalah fase pada mikrofon ganda atau untuk efek tertentu.

### B.2 Operasi Penjumlahan Citra

Kita akan melakukan operasi penjumlahan dua citra: `I(i, j) = I1(i, j) + I2(i, j)`.
Pastikan kedua citra memiliki ukuran yang sama. Jika ukuran berbeda, akan dilakukan *resizing* terlebih dahulu. Kita akan menggunakan `image` yang sudah ada sebagai `image1` dan memuat citra lain sebagai `image2`.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage import data, transform  # transform digunakan untuk mengubah ukuran citra

# Menggunakan citra pertama dari variabel yang sudah ada
image1 = image.copy()

# Mengambil citra kedua dari library scikit-image
image2_raw = data.astronaut()

# Memastikan kedua citra memiliki
# jumlah channel yang sama

# Jika citra pertama grayscale, ubah menjadi RGB
if len(image1.shape) == 2:
    image1 = cv2.cvtColor(image1, cv2.COLOR_GRAY2RGB)

# Jika citra kedua grayscale, ubah menjadi RGB
if len(image2_raw.shape) == 2:
    image2_raw = cv2.cvtColor(image2_raw, cv2.COLOR_GRAY2RGB)

# Mengambil ukuran kedua citra
h1, w1 = image1.shape[:2]
h2, w2 = image2_raw.shape[:2]

# Variabel untuk menyimpan penjelasan resizing
resizing_explanation = ""

# Menyesuaikan ukuran citra

if h1 != h2 or w1 != w2:

    # Mengubah ukuran citra kedua agar sama dengan citra pertama
    image2 = transform.resize(
        image2_raw,
        (h1, w1),
        anti_aliasing=True
    )

    # Normalisasi nilai piksel ke rentang 0–1
    image1 = image1.astype(np.float32) / 255.0

    resizing_explanation = (
        f"Citra kedua diubah ukurannya dari "
        f"{h2}x{w2} menjadi {h1}x{w1} "
        f"agar sesuai dengan citra pertama."
    )

else:

    # Jika ukuran sama, langsung normalisasi
    image2 = image2_raw.astype(np.float32) / 255.0
    image1 = image1.astype(np.float32) / 255.0

# Operasi Penjumlahan Citra

# Menjumlahkan kedua citra piksel demi piksel
added_image_float = image1 + image2

# Metode Clipping

# Membatasi nilai piksel agar tetap berada pada rentang 0–1
added_image_clipped = np.clip(
    added_image_float,
    0,
    1
)

# Metode Normalisasi

# Mencari nilai maksimum piksel
max_val = np.max(added_image_float)

if max_val > 0:
    added_image_normalized = added_image_float / max_val
else:
    added_image_normalized = added_image_float

# Mengubah kembali ke format uint8

image1_display = (image1 * 255).astype(np.uint8)
image2_display = (image2 * 255).astype(np.uint8)

added_image_clipped_display = (
    added_image_clipped * 255
).astype(np.uint8)

added_image_normalized_display = (
    added_image_normalized * 255
).astype(np.uint8)

# Menampilkan hasil

plt.figure(figsize=(18, 6))

plt.subplot(1, 4, 1)
plt.imshow(image1_display)
plt.title('Citra Pertama (Chelsea)')
plt.axis('off')

plt.subplot(1, 4, 2)
plt.imshow(image2_display)
plt.title('Citra Kedua (Astronaut)')
plt.axis('off')

plt.subplot(1, 4, 3)
plt.imshow(added_image_clipped_display)
plt.title('Hasil Penjumlahan (Clipping)')
plt.axis('off')

plt.subplot(1, 4, 4)
plt.imshow(added_image_normalized_display)
plt.title('Hasil Penjumlahan (Normalisasi)')
plt.axis('off')

plt.tight_layout()
plt.show()

# Menampilkan informasi resizing

print("\nPenjelasan proses resizing:")

if resizing_explanation:
    print(resizing_explanation)
else:
    print(
        "Tidak dilakukan resizing karena "
        "ukuran kedua citra sudah sama."
    )
```

### Analisis Efek Penjumlahan terhadap Brightness dan Detail Citra

Operasi penjumlahan citra `I(i, j) = I1(i, j) + I2(i, j)` secara umum akan meningkatkan nilai piksel. Hal ini berdampak pada:

*   **Brightness (Kecerahan):** Citra hasil penjumlahan cenderung akan **lebih cerah** dibandingkan kedua citra asalnya. Ini karena nilai piksel dari kedua citra ditambahkan. Jika piksel kedua citra memiliki nilai tinggi pada lokasi yang sama, maka piksel hasil penjumlahan akan memiliki nilai yang sangat tinggi, bahkan bisa melebihi rentang maksimum tipe data (misalnya 255 untuk `uint8`). Inilah mengapa penanganan *overflow* (seperti *clipping* atau *normalisasi*) menjadi penting.
*   **Detail:** Detail dari kedua citra akan **tergabung** ke dalam citra hasil penjumlahan. Area yang terang di salah satu citra dan gelap di citra lain akan memiliki nilai menengah. Area dengan detail yang jelas di kedua citra akan tetap memiliki detail yang jelas, namun mungkin dengan kecerahan yang meningkat. Pada area yang gelap di kedua citra, detail mungkin masih terlihat samar.

Jika terjadi *clipping*, detail pada area yang terlalu terang bisa hilang karena semua nilai di atas ambang batas maksimum dipotong ke nilai maksimum tersebut. Normalisasi mempertahankan rentang dinamis, tetapi dapat mengurangi kontras keseluruhan jika ada beberapa piksel yang sangat terang yang 'mendorong' semua piksel lain ke bawah setelah penskalaan.

### Pertanyaan Analisis

1.  **Mengapa ukuran citra harus sama sebelum dijumlahkan?**
    Ukuran citra harus sama sebelum dijumlahkan karena operasi penjumlahan citra dilakukan secara **piksel-demi-piksel** pada posisi (i, j) yang sama. Jika ukuran citra berbeda, tidak ada korespondensi satu-satu antar piksel, sehingga operasi penjumlahan tidak dapat didefinisikan secara langsung untuk setiap piksel. Mencoba menjumlahkan citra dengan ukuran berbeda tanpa penyesuaian akan menghasilkan kesalahan dimensi atau hasil yang tidak bermakna.

2.  **Apa yang terjadi jika hasil penjumlahan melebihi nilai maksimum piksel?**
    Jika hasil penjumlahan melebihi nilai maksimum piksel (misalnya, `255` untuk citra `uint8`), akan terjadi **overflow**. Ada dua cara umum penanganannya:
    *   **Clipping (Penjepitan):** Nilai piksel yang melebihi batas maksimum akan dipotong ke nilai maksimum tersebut (misalnya, `260` menjadi `255`). Ini menjaga nilai dalam rentang yang valid tetapi dapat menyebabkan hilangnya detail pada area yang terlalu terang (disebut *saturation* atau *blown-out highlights*).
    *   **Wraparound (Pembungkusan):** (Kurang umum, tetapi bisa terjadi pada operasi aritmetika integer jika tidak ditangani secara eksplisit) Nilai akan 'membungkus' kembali dari batas bawah. Misalnya, `256` bisa menjadi `0`, `257` menjadi `1`, dll. Ini menghasilkan artefak visual yang tidak diinginkan dan mengubah warna secara drastis.

3.  **Apa perbedaan hasil jika menggunakan *clipping* dan *normalisasi*?**
    *   **Clipping:** Setelah penjumlahan, nilai piksel yang melebihi batas atas akan dipotong ke batas atas tersebut, dan nilai yang kurang dari batas bawah akan dipotong ke batas bawah. Ini menjaga kecerahan relatif pada sebagian besar rentang, tetapi mengorbankan detail pada area yang sangat terang. Citra mungkin terlihat lebih kontras tetapi kehilangan informasi di *highlight*.
    *   **Normalisasi:** Seluruh rentang nilai piksel hasil penjumlahan akan diskalakan ulang agar pas dalam rentang yang diinginkan (misalnya `[0, 255]`). Ini mempertahankan semua detail dan rentang dinamis citra, tetapi dapat membuat citra terlihat kurang kontras secara keseluruhan jika ada beberapa piksel yang sangat terang, karena piksel-piksel tersebut akan 'meregangkan' rentang dan membuat piksel lainnya menjadi relatif lebih gelap.

4.  **Dalam aplikasi nyata, operasi ini dapat digunakan untuk apa?**
    Operasi penjumlahan citra memiliki beberapa aplikasi nyata:
    *   **Image Blending/Layering:** Menggabungkan beberapa gambar (misalnya, menambahkan logo, *watermark*, atau efek khusus ke sebuah foto).
    *   **Noise Reduction (dengan averaging):** Jika kita memiliki beberapa citra yang sama dengan *noise* acak yang berbeda, menjumlahkan (dan kemudian merata-ratakan) citra-citra tersebut dapat mengurangi *noise* dan meningkatkan kualitas gambar.
    *   **Exposure Stacking:** Menggabungkan beberapa citra dengan eksposur berbeda untuk menciptakan gambar dengan rentang dinamis yang lebih luas (HDR).
    *   **Background Subtraction:** Dalam pengolahan video, penjumlahan (atau pengurangan) dapat digunakan untuk mendeteksi perubahan atau objek yang bergerak dengan membandingkan frame saat ini dengan frame latar belakang.
    *   **Medical Imaging:** Dalam citra medis, penjumlahan dapat digunakan untuk menggabungkan informasi dari modalitas pencitraan yang berbeda atau untuk meningkatkan rasio sinyal terhadap *noise*.

### B.3 Operasi Penggeseran Citra

Kita akan melakukan translasi (penggeseran) citra menggunakan persamaan: `I′(i, j) = I(i − Δi, j − Δj)`.
Kita akan menggunakan citra `image1` (Chelsea) yang sudah di-load sebelumnya dan menerapkan minimal tiga kombinasi penggeseran:
*   `Δi = 30, Δj = 0` (geser ke bawah)
*   `Δi = 0, Δj = 30` (geser ke kanan)
*   `Δi = 30, Δj = 30` (geser ke bawah dan kanan)

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Menggunakan citra asli yang telah dimuat sebelumnya
# Variabel image berasal dari praktikum sebelumnya
image_original = image.copy()

# Menentukan nilai translasi
# (Δi = pergeseran vertikal,
#  Δj = pergeseran horizontal)

translation_values = [
    (30, 0),
    (0, 30),
    (30, 30)
]

# Menyimpan hasil translasi
shifted_images = {}

# Menampilkan citra asli

plt.figure(figsize=(15, 5))

plt.subplot(1, len(translation_values) + 1, 1)
plt.imshow(image_original)
plt.title('Citra Asli')
plt.axis('off')

# Melakukan translasi citra

for i, (delta_i, delta_j) in enumerate(translation_values):

    # Membuat matriks translasi
    # [1 0 tx]
    # [0 1 ty]
    M = np.float32([
        [1, 0, delta_j],
        [0, 1, delta_i]
    ])

    # Mengambil ukuran citra
    rows, cols = image_original.shape[:2]

    # Melakukan transformasi translasi
    shifted_image = cv2.warpAffine(
        image_original,
        M,
        (cols, rows),
        borderValue=(0, 0, 0)  # Area kosong diisi warna hitam
    )

    # Menyimpan hasil translasi
    shifted_images[(delta_i, delta_j)] = shifted_image

    # Menampilkan hasil translasi
    plt.subplot(
        1,
        len(translation_values) + 1,
        i + 2
    )

    plt.imshow(shifted_image)
    plt.title(
        f'Geser Δi={delta_i}, Δj={delta_j}'
    )
    plt.axis('off')

# Mengatur tata letak gambar
plt.tight_layout()
plt.show()

# Menampilkan penjelasan

print("\nPenjelasan area kosong setelah penggeseran:")

print(
    "Ketika citra digeser, bagian citra yang 'keluar' dari bingkai asli akan hilang. Sebaliknya, area kosong akan muncul di sisi yang berlawanan dari arah pergeseran. Area kosong ini biasanya diisi dengan nilai piksel nol (hitam) secara default oleh fungsi cv2.warpAffine. Misalnya, jika citra digeser ke kanan dan ke bawah, area kosong hitam akan muncul di sisi kiri dan atas citra, karena tidak ada informasi piksel dari citra asli yang mengisi area tersebut."
)
```

### Pertanyaan Analisis

1.  **Apa efek penggeseran terhadap posisi objek dalam citra?**
    Efek penggeseran (translasi) adalah memindahkan posisi objek dalam citra dari satu lokasi ke lokasi lain tanpa mengubah ukuran, orientasi, atau bentuk objek. Jika citra digeser sejauh `(Δi, Δj)`, maka setiap piksel `(i, j)` di citra asli akan berpindah ke posisi `(i + Δi, j + Δj)` di citra hasil geser. Artinya, seluruh isi citra, termasuk objek-objek di dalamnya, bergeser secara kolektif sesuai dengan vektor translasi yang diberikan.

2.  **Mengapa translasi digunakan dalam augmentasi data?**
    Translasi sering digunakan dalam augmentasi data (data augmentation) untuk *machine learning* dan *deep learning* karena alasan berikut:
    *   **Meningkatkan Robustness Model:** Dengan menghasilkan versi citra yang digeser dari dataset pelatihan, model belajar untuk mengenali objek terlepas dari posisi mereka dalam citra. Ini membuat model lebih *robust* (tangguh) terhadap variasi posisi objek di dunia nyata.
    *   **Memperluas Dataset:** Translasi secara efektif memperluas ukuran dataset pelatihan tanpa perlu mengumpulkan data baru secara manual, yang bisa mahal dan memakan waktu.
    *   **Mencegah Overfitting:** Dengan memperkenalkan variasi pada data pelatihan, augmentasi data membantu mencegah model untuk terlalu spesifik pada contoh-contoh pelatihan asli, sehingga meningkatkan kemampuan generalisasi model pada data yang belum pernah dilihat.

3.  **Apa risiko penggunaan penggeseran yang terlalu besar pada citra?**
    Penggunaan penggeseran yang terlalu besar pada citra memiliki beberapa risiko:
    *   **Kehilangan Informasi Penting:** Objek utama atau bagian penting dari citra dapat tergeser keluar dari bingkai, menyebabkan hilangnya informasi krusial. Misalnya, jika wajah seseorang digeser terlalu jauh, sebagian besar wajah mungkin tidak lagi terlihat.
    *   **Generasi Data Tidak Realistis:** Penggeseran yang ekstrem dapat menghasilkan citra yang tidak realistis atau tidak merepresentasikan variasi yang mungkin terjadi di dunia nyata, yang justru dapat membingungkan model *machine learning* dan mengurangi performanya.
    *   **Area Kosong yang Besar:** Pergeseran besar akan menciptakan area kosong (biasanya hitam) yang signifikan di citra, yang dapat memperkenalkan *noise* atau fitur yang tidak relevan jika tidak ditangani dengan benar oleh model.

### B.4 Operasi Amplifikasi Citra

Kita akan melakukan operasi amplifikasi citra: `I′(i, j) = αI(i, j)`.
Kita akan menggunakan citra `image_original` (Chelsea) yang sudah di-load sebelumnya dan mengamplifikasinya dengan minimal empat nilai `α`: `0.5`, `1`, `1.5`, dan `2`.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Menggunakan citra asli dari praktikum sebelumnya
# Mengubah tipe data menjadi float dan menormalisasi ke rentang 0–1
image_float = image_original.astype(np.float32) / 255.0

# Menentukan faktor amplifikasi (α)

alpha_values = [0.5, 1, 1.5, 2]

# Menyimpan hasil amplifikasi
amplified_images = {}

# Melakukan amplifikasi citra

for alpha in alpha_values:

    # Mengalikan setiap piksel dengan faktor α
    amplified_float = alpha * image_float

    # Membatasi nilai piksel agar tetap berada pada rentang 0–1
    amplified_clipped_float = np.clip(
        amplified_float,
        0,
        1
    )

    # Mengubah kembali ke format uint8
    amplified_images[alpha] = (
        amplified_clipped_float * 255
    ).astype(np.uint8)

# ==================================
# Menampilkan citra hasil amplifikasi
# ==================================

plt.figure(figsize=(15, 8))

# Citra asli
plt.subplot(1, len(alpha_values) + 1, 1)
plt.imshow(image_original)
plt.title('Citra Asli')
plt.axis('off')

# Citra hasil amplifikasi
for i, alpha in enumerate(alpha_values):

    plt.subplot(
        1,
        len(alpha_values) + 1,
        i + 2
    )

    plt.imshow(amplified_images[alpha])
    plt.title(f'Amplifikasi α = {alpha}')
    plt.axis('off')

plt.tight_layout()
plt.show()

# Menampilkan histogram

plt.figure(figsize=(15, 8))

# Histogram citra asli
plt.subplot(2, len(alpha_values) + 1, 1)

plt.title('Histogram Citra Asli')
plt.hist(
    image_original.ravel(),
    bins=256,
    range=[0, 256],
    color='gray',
    histtype='stepfilled'
)

plt.xlabel('Nilai Piksel')
plt.ylabel('Frekuensi')

# Menyamakan batas sumbu y
plt.ylim(
    0,
    np.max(
        plt.hist(
            image_original.ravel(),
            bins=256,
            range=[0, 256]
        )[0]
    ) * 1.2
)

# Histogram hasil amplifikasi
for i, alpha in enumerate(alpha_values):

    plt.subplot(
        2,
        len(alpha_values) + 1,
        i + 2
    )

    plt.title(f'Histogram α = {alpha}')

    plt.hist(
        amplified_images[alpha].ravel(),
        bins=256,
        range=[0, 256],
        color='gray',
        histtype='stepfilled'
    )

    plt.xlabel('Nilai Piksel')
    plt.ylabel('Frekuensi')

    plt.ylim(
        0,
        np.max(
            plt.hist(
                image_original.ravel(),
                bins=256,
                range=[0, 256]
            )[0]
        ) * 1.2
    )

plt.tight_layout()
plt.show()

# Analisis hasil

print("\nAnalisis perubahan brightness dan kontras:")

print(
    "Ketika α > 1, nilai piksel meningkat sehingga "
    "citra menjadi lebih terang. Kontras juga dapat "
    "meningkat karena perbedaan intensitas antar piksel "
    "menjadi lebih besar."
)

print(
    "Ketika 0 < α < 1, nilai piksel menurun sehingga "
    "citra menjadi lebih gelap dan kontras cenderung "
    "berkurang."
)

print(
    "Histogram menunjukkan distribusi nilai piksel. "
    "Jika α > 1, histogram bergeser ke kanan "
    "(nilai piksel lebih tinggi). "
    "Jika α < 1, histogram bergeser ke kiri "
    "(nilai piksel lebih rendah)."
)

print(
    "Apabila terjadi clipping, sejumlah piksel akan "
    "menumpuk pada nilai maksimum (255), sehingga "
    "detail pada area terang dapat hilang."
)
```

### Pertanyaan Analisis

1.  **Apa efek α > 1 pada citra?**
    Ketika `α > 1`, citra menjadi **lebih cerah (brighter)** karena setiap nilai piksel dikalikan dengan faktor lebih besar dari 1. Amplitudo (intensitas) setiap piksel meningkat. Ini juga dapat menyebabkan **peningkatan kontras** jika tidak ada *clipping* yang signifikan, namun jika nilai piksel melebihi batas maksimum (misalnya 255), maka *clipping* akan terjadi dan detail pada area yang terang akan hilang (menjadi *saturated* atau *blown out*).

2.  **Apa efek 0 < α < 1 pada citra?**
    Ketika `0 < α < 1`, citra menjadi **lebih gelap (dimmer)** karena setiap nilai piksel dikalikan dengan faktor kurang dari 1. Amplitudo (intensitas) setiap piksel menurun. Ini biasanya menyebabkan **penurunan kontras** secara keseluruhan, membuat citra terlihat lebih 'flat' atau kurang dinamis, karena perbedaan antara nilai piksel terang dan gelap menjadi lebih kecil.

3.  **Mengapa nilai piksel perlu dibatasi pada rentang tertentu?**
    Nilai piksel perlu dibatasi pada rentang tertentu (misalnya `[0, 255]` untuk citra `uint8` atau `[0, 1]` untuk citra *floating-point*) karena:
    *   **Representasi Data:** Setiap format citra digital memiliki batasan bit untuk menyimpan informasi piksel. Misalnya, `uint8` menggunakan 8 bit, yang hanya dapat menyimpan nilai integer dari 0 hingga 255.
    *   **Tampilan:** Perangkat tampilan (monitor, printer) hanya dapat menampilkan rentang intensitas cahaya tertentu. Nilai di luar rentang ini tidak dapat direpresentasikan secara fisik dan akan menyebabkan *clipping*.
    *   **Pencegahan Overflow/Underflow:** Pembatasan mencegah masalah *overflow* (nilai terlalu tinggi) dan *underflow* (nilai terlalu rendah) yang dapat menyebabkan distorsi warna, artefak, atau data yang tidak valid saat operasi aritmetika dilakukan.

4.  **Apa hubungan amplifikasi citra dengan peningkatan brightness?**
    Amplifikasi citra secara langsung berhubungan dengan peningkatan *brightness*. Ketika sebuah citra diamplifikasi dengan faktor `α > 1`, setiap nilai intensitas piksel `I(i, j)` dikalikan dengan `α` menjadi `αI(i, j)`. Jika `I(i, j)` merepresentasikan kecerahan, maka mengalikan dengan `α` yang lebih besar dari 1 akan meningkatkan nilai kecerahan tersebut, membuat piksel (dan keseluruhan citra) tampak lebih terang. Sebaliknya, `α` antara 0 dan 1 akan mengurangi nilai kecerahan, membuat citra lebih gelap.

### C.2 Uji Additivitas

Kita akan menggunakan sistem yang sama, yaitu `T(x) = 2x`.

Sifat additivitas menyatakan bahwa:
`T(x1 + x2) = T(x1) + T(x2)`

Kita akan menguji properti ini menggunakan sinyal `x1[n]` dan `x2[n]` yang sudah kita definisikan di Bagian A.1.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definisikan sistem T(x) = 2x
def T(signal):
    return 2 * signal

# Gunakan sinyal x1[n] dan x2[n] dari Bagian A.1
# n, x1, x2 sudah didefinisikan sebelumnya

# 1. Plot x1[n]
plt.figure(figsize=(12, 4))
plt.stem(n, x1)
plt.title('Sinyal Input x1[n]')
plt.xlabel('n')
plt.ylabel('x1[n]')
plt.grid(True)
plt.xticks(n[::5])
plt.tight_layout()
plt.show()

# 2. Plot x2[n]
plt.figure(figsize=(12, 4))
plt.stem(n, x2)
plt.title('Sinyal Input x2[n]')
plt.xlabel('n')
plt.ylabel('x2[n]')
plt.grid(True)
plt.xticks(n[::5])
plt.tight_layout()
plt.show()

# Hitung T(x1 + x2)
x1_plus_x2 = x1 + x2
T_x1_plus_x2 = T(x1_plus_x2)

# Hitung T(x1) + T(x2)
T_x1 = T(x1)
T_x2 = T(x2)
T_x1_plus_T_x2 = T_x1 + T_x2

print("\nPengujian Additivitas untuk T(x) = 2x:")
print(f"  T(x1 + x2) = {T_x1_plus_x2[0]:.2f}, ..., {T_x1_plus_x2[-1]:.2f}")
print(f"  T(x1) + T(x2) = {T_x1_plus_T_x2[0]:.2f}, ..., {T_x1_plus_T_x2[-1]:.2f}")

# 3. Plot T(x1 + x2)
plt.figure(figsize=(12, 5))
plt.stem(n, T_x1_plus_x2)
plt.title('Hasil T(x1 + x2)')
plt.xlabel('n')
plt.ylabel('Amplitudo')
plt.grid(True)
plt.xticks(n[::5])
plt.tight_layout()
plt.show()

# 4. Plot T(x1) + T(x2)
plt.figure(figsize=(12, 5))
plt.stem(n, T_x1_plus_T_x2)
plt.title('Hasil T(x1) + T(x2)')
plt.xlabel('n')
plt.ylabel('Amplitudo')
plt.grid(True)
plt.xticks(n[::5])
plt.tight_layout()
plt.show()

# 5. Plot Selisih antara kedua hasil
difference = T_x1_plus_x2 - T_x1_plus_T_x2
plt.figure(figsize=(12, 4))
plt.stem(n, difference)
plt.title('Selisih: T(x1 + x2) - (T(x1) + T(x2))')
plt.xlabel('n')
plt.ylabel('Selisih')
plt.grid(True)
plt.xticks(n[::5])
plt.tight_layout()
plt.show()

# 6. Kesimpulan
if np.allclose(T_x1_plus_x2, T_x1_plus_T_x2):
    print("\n-> T(x1 + x2) SAMA dengan T(x1) + T(x2). Sistem MEMENUHI sifat additivitas.")
else:
    print("\n-> T(x1 + x2) TIDAK SAMA dengan T(x1) + T(x2). Sistem TIDAK MEMENUHI sifat additivitas.")
```

### C.3 Bandingkan Sistem Linier dan Non-Linier

Kita akan membandingkan dua sistem:
1.  **Sistem Linier:** `T1(x) = 2x`
2.  **Sistem Non-Linier:** `T2(x) = x^2`

Kita akan menguji kedua sistem ini menggunakan sifat homogenitas dan additivitas.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Definisikan sistem T1(x) = 2x
def T1(signal):
    return 2 * signal

# Definisikan sistem T2(x) = x^2
def T2(signal):
    return signal**2

# Gunakan sinyal x1[n] dan x2[n] dari Bagian A.1
# n, x1, x2 sudah didefinisikan sebelumnya
x_input = x1 # Sinyal tunggal untuk homogenitas

# --- Uji Homogenitas ---
alpha_values = [0.5, 2] # Contoh nilai alpha

results_homogeneity = []

# Uji T1(x) = 2x untuk homogenitas
print("\n--- Uji Homogenitas untuk T1(x) = 2x ---")
for alpha in alpha_values:
    T1_alpha_x = T1(alpha * x_input)
    alpha_T1_x = alpha * T1(x_input)
    is_homogeneous = np.allclose(T1_alpha_x, alpha_T1_x)
    print(f"  α = {alpha}: T1(αx) {'==' if is_homogeneous else '!='} αT1(x) -> {is_homogeneous}")
    results_homogeneity.append({'Sistem': 'T1(x) = 2x', 'Alpha': alpha, 'T(αx)': T1_alpha_x, 'αT(x)': alpha_T1_x, 'Homogen': is_homogeneous})

# Uji T2(x) = x^2 untuk homogenitas
print("\n--- Uji Homogenitas untuk T2(x) = x^2 ---")
for alpha in alpha_values:
    T2_alpha_x = T2(alpha * x_input)
    alpha_T2_x = alpha * T2(x_input)
    is_homogeneous = np.allclose(T2_alpha_x, alpha_T2_x)
    print(f"  α = {alpha}: T2(αx) {'==' if is_homogeneous else '!='} αT2(x) -> {is_homogeneous}")
    results_homogeneity.append({'Sistem': 'T2(x) = x^2', 'Alpha': alpha, 'T(αx)': T2_alpha_x, 'αT(x)': alpha_T2_x, 'Homogen': is_homogeneous})

# --- Uji Additivitas ---
results_additivity = []

# Uji T1(x) = 2x untuk additivitas
print("\n--- Uji Additivitas untuk T1(x) = 2x ---")
T1_x1_plus_x2 = T1(x1 + x2)
T1_x1_plus_T1_x2 = T1(x1) + T1(x2)
is_additive_T1 = np.allclose(T1_x1_plus_x2, T1_x1_plus_T1_x2)
print(f"  T1(x1 + x2) {'==' if is_additive_T1 else '!='} T1(x1) + T1(x2) -> {is_additive_T1}")
results_additivity.append({'Sistem': 'T1(x) = 2x', 'T(x1+x2)': T1_x1_plus_x2, 'T(x1)+T(x2)': T1_x1_plus_T1_x2, 'Aditif': is_additive_T1})

# Uji T2(x) = x^2 untuk additivitas
print("\n--- Uji Additivitas untuk T2(x) = x^2 ---")
T2_x1_plus_x2 = T2(x1 + x2)
T2_x1_plus_T2_x2 = T2(x1) + T2(x2)
is_additive_T2 = np.allclose(T2_x1_plus_x2, T2_x1_plus_T2_x2)
print(f"  T2(x1 + x2) {'==' if is_additive_T2 else '!='} T2(x1) + T2(x2) -> {is_additive_T2}")
results_additivity.append({'Sistem': 'T2(x) = x^2', 'T(x1+x2)': T2_x1_plus_x2, 'T(x1)+T(x2)': T2_x1_plus_T2_x2, 'Aditif': is_additive_T2})

# --- Tabel Perbandingan ---

comparison_table = pd.DataFrame([
    {'Sistem': 'T1(x) = 2x', 'Homogenitas': results_homogeneity[0]['Homogen'], 'Additivitas': results_additivity[0]['Aditif']},
    {'Sistem': 'T2(x) = x^2', 'Homogenitas': results_homogeneity[2]['Homogen'], 'Additivitas': results_additivity[1]['Aditif']}
])

comparison_table['Linier / Tidak Linier'] = np.where(
    (comparison_table['Homogenitas'] == True) & (comparison_table['Additivitas'] == True),
    'Linier', 'Tidak Linier'
)

comparison_table['Alasan'] = np.where(
    comparison_table['Sistem'] == 'T1(x) = 2x',
    'Memenuhi homogenitas dan additivitas.',
    'Tidak memenuhi homogenitas dan additivitas.'
)

print("\n--- Tabel Perbandingan Sistem Linier dan Non-Linier ---")
print(comparison_table.to_markdown(index=False))
```

### Pertanyaan Analisis

1.  **Mengapa T1(x) = 2x dapat disebut sistem linier?**
    Sistem `T1(x) = 2x` disebut sistem linier karena memenuhi kedua sifat sistem linier:
    *   **Homogenitas:** `T1(αx) = 2(αx) = α(2x) = αT1(x)`. Outputnya sebanding dengan input yang diskalakan.
    *   **Additivitas:** `T1(x1 + x2) = 2(x1 + x2) = 2x1 + 2x2 = T1(x1) + T1(x2)`. Output dari penjumlahan input sama dengan penjumlahan output masing-masing input.
    Karena kedua sifat ini terpenuhi, `T1(x) = 2x` adalah sistem linier.

2.  **Mengapa T2(x) = x^2 bukan sistem linier?**
    Sistem `T2(x) = x^2` bukan sistem linier karena tidak memenuhi kedua sifat sistem linier:
    *   **Homogenitas:** `T2(αx) = (αx)^2 = α^2 x^2`. Sedangkan `αT2(x) = α(x^2)`. Karena `α^2 x^2 ≠ αx^2` (kecuali `α=0` atau `α=1`), sifat homogenitas tidak terpenuhi.
    *   **Additivitas:** `T2(x1 + x2) = (x1 + x2)^2 = x1^2 + 2x1x2 + x2^2`. Sedangkan `T2(x1) + T2(x2) = x1^2 + x2^2`. Karena `x1^2 + 2x1x2 + x2^2 ≠ x1^2 + x2^2` (kecuali `x1=0` atau `x2=0`), sifat additivitas tidak terpenuhi.
    Oleh karena itu, `T2(x) = x^2` adalah sistem non-linier.

3.  **Mengapa konsep sistem linier penting dalam Pengolahan Sinyal Digital?**
    Konsep sistem linier sangat penting dalam Pengolahan Sinyal Digital (PSD) karena:
    *   **Analisis yang Lebih Mudah:** Sistem linier jauh lebih mudah untuk dianalisis secara matematis. Banyak alat analisis yang kuat (seperti Transformasi Fourier, Transformasi Laplace, dan Transformasi-Z) dikembangkan khusus untuk sistem linier.
    *   **Prinsip Superposisi:** Sifat superposisi (gabungan homogenitas dan additivitas) memungkinkan kita untuk menganalisis respons sistem terhadap input yang kompleks dengan memecahnya menjadi input yang lebih sederhana dan menjumlahkan respons individualnya. Ini sangat menyederhanakan desain filter, ekualiser, dan sistem PSD lainnya.
    *   **Desain Filter:** Hampir semua filter digital (FIR, IIR) dirancang sebagai sistem linier karena sifat-sifat ini menjamin bahwa filter akan mengubah sinyal input dengan cara yang dapat diprediksi (misalnya, hanya mengubah amplitudo dan fase, tidak menciptakan frekuensi baru).
    *   **Prediksi Perilaku:** Dalam sistem linier, kita dapat memprediksi output untuk setiap input jika kita mengetahui respons impuls (atau respons frekuensi) sistem. Ini tidak mungkin dilakukan pada sistem non-linier secara umum.
    *   **Efisiensi Komputasi:** Banyak operasi PSD yang efisien (misalnya, konvolusi) memanfaatkan sifat linieritas.

4.  **Apa hubungan sistem linier dengan superposisi?**
    Hubungan antara sistem linier dan prinsip superposisi adalah **identik**. Sebuah sistem disebut linier **jika dan hanya jika** sistem tersebut memenuhi prinsip superposisi. Prinsip superposisi sendiri adalah kombinasi dari dua sifat utama:
    *   **Homogenitas (Skalabilitas):** Jika input diskalakan dengan faktor `α`, maka output juga diskalakan dengan faktor `α` yang sama (`T(αx) = αT(x)`).
    *   **Additivitas:** Jika dua input dijumlahkan, maka outputnya adalah penjumlahan dari output masing-masing input secara terpisah (`T(x1 + x2) = T(x1) + T(x2)`).
    Jadi, ketika kita mengatakan sebuah sistem adalah linier, kita secara implisit menyatakan bahwa sistem tersebut mematuhi prinsip superposisi. Prinsip ini adalah landasan teoritis yang memungkinkan pengembangan banyak teknik dan algoritma penting dalam pengolahan sinyal.
